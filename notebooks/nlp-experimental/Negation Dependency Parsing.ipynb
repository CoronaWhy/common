```python
import pandas as pd
import scispacy
import spacy
from spacy import displacy
```

```python
# Load the data from the first sheet
df = pd.read_excel("/home/acorn/Downloads/DrugVisData_Negated_Output_Parsed_Validations.xlsx")
```

```python
# Load our language model for dependency parsing
nlp = spacy.load("/home/acorn/Downloads/en_core_sci_lg-0.2.4/en_core_sci_lg/en_core_sci_lg-0.2.4/")
```

```python
# Turn all our sentences int parsed spaCy documents
docs = [nlp(str(sentence)) for sentence in df["sentence"].to_list()]
```

```python
# Take a look at how dependencies are visualized/parsed. 
displacy.serve(docs[0], style="dep")
```

```python
# Iterate over dataframe to find if any drugs are under a NEG head in the dependency parse.
for i in range(len(df.index)):
    negation_tokens = [tok for tok in docs[i] if tok.dep_ == 'neg' and tok.text == str(df.iloc[i]["drug"])]
    negation_head_tokens = [token.head for token in negation_tokens]

    for token in negation_head_tokens:
        print(token.text, token.dep_, token.head.text, token.head.pos_, [child for child in token.children])
        
# Please note this is super basic. 
```
